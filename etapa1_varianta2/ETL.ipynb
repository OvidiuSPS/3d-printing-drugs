```python
import pandas as pd
```

```python
# Specify the path to your Excel file
file_path = 'Etapa1 - completat 11.2024.xlsx'

# Read all sheets into a dictionary of DataFrames
all_sheets = pd.read_excel(file_path, sheet_name=None)


# Iterate through the dictionary of DataFrames
for sheet_name, df in all_sheets.items():
    # Add a new column with the sheet name
    df['SheetName'] = sheet_name
    
    # Assign the DataFrame to a variable named after the sheet name
    globals()[sheet_name] = df
    globals()[sheet_name].set_index('Primary ID', inplace=True)
```

```python
cols_prof_cedare = ["%5min", "%10min", "%15min", "%30min", "%60min", "%120min", "%180min", "%240min"]
x = globals()['Input1_Etapa1'].drop(columns=cols_prof_cedare)
compozitie_df = globals()['Output_Etapa1']
y = globals()['Input1_Etapa1'][cols_prof_cedare]
X = x.merge(compozitie_df, left_index=True, right_index=True)
```

```python
# List of columns to drop
columns_to_drop = ['Sec ID', 'SheetName_x', "SheetName_y"]

# Drop the columns only if they exist in the DataFrame
X = X.drop(columns=[col for col in columns_to_drop if col in X.columns], errors='ignore')
y = y.drop(columns=[col for col in columns_to_drop if col in y.columns], errors='ignore')

# Get the indices of rows containing NaN values in df_X
nan_indices = X[X.isna().any(axis=1)].index
print(f"Indices of NaN rows: {nan_indices}")

# Remove rows with these indices in both df_X and df_Y
X.drop(nan_indices, inplace=True)
y.drop(nan_indices, inplace=True)

# Check the first few rows after dropping the columns
print(X.head())
print(y.head())

# check nr de linii
assert X.shape[0] == y.shape[0]
```

```python
# Identify rows with NaNs
data = pd.merge(X, y, left_index=True, right_index=True)

rows_with_nans = data.isnull().any(axis=1)
num_rows_with_nans = rows_with_nans.sum()
total_rows = len(data)
fraction_rows_with_nans = num_rows_with_nans / total_rows

print(f"Number of rows with NaNs: {num_rows_with_nans}")
print(f"Fraction of rows with NaNs: {fraction_rows_with_nans:.2f}")

# Identify columns with NaNs
columns_with_nans = data.columns[data.isnull().any()].tolist()
num_columns_with_nans = len(columns_with_nans)

print(f"Columns with NaNs: {columns_with_nans}")

# Print detailed information about NaNs in each column
nan_info = data.isnull().sum()
print("\nDetailed NaN information:")
print(nan_info[nan_info > 0])
```
